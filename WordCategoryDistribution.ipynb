```python
import nltk
import csv
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

while True: 
    # Provide some text in English
    sentence = input("Provide some text in English: ")
    # Tokenize the sentence (split into words)
    rawtokens = nltk.word_tokenize(sentence)
    # Define punctuation to remove
    punctuation = [".", ",", ";", ":", "'", '"', "!", "?","[","]","(",")",'-']
    # Remove punctuation from rawtokens
    filtered_tokens = [token for token in rawtokens if token not in punctuation]
    # Get the category for each word (using filtered rawtokens)
    pos_tags = nltk.pos_tag(filtered_tokens, tagset='universal')

    # Print the word and its category (Part-of-Speech)
    for token, tag in pos_tags: 
        print(f"{token}: {tag}")


    # Store the whole information in a CSV file. 
    with open('WordsByTag.csv', 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['Word', 'Tag']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
    
    # Write each word and its category to the CSV
        for word, tag in pos_tags:
            writer.writerow({'Word': word, 'Tag': tag})
            
    # Create a Counter object to store tag occurrences
    tag_counts = Counter()
    for _, tag in pos_tags:
        tag_counts[tag]+= 1
    # Store tag occurrences in a CSV file
    with open('TagCounts.csv', 'w', newline='') as csvfile:
        fieldnames = ['Tag', 'Count']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        # Sort tag_counts by frequency in descending order (highest first)
        sorted_counts = sorted(tag_counts.items(), key=lambda item: item[1], reverse=True)
        # Write each tag and its count as a dictionary row
        for tag, count in sorted_counts:
            writer.writerow({'Tag': tag, 'Count': count})
            
   
    # Create a bar chart describing the Part-of-Speach distribution in the given text 
    plt.figure(figsize=(8, 6))  
    plt.bar(tag_counts.keys(), tag_counts.values(), color='skyblue')  
    plt.xlabel("Part-of-Speech Tag")
    plt.ylabel("Number of Occurrences")
    plt.title("Part-of-Speech Tag Distribution")
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
    plt.tight_layout()  # Adjust layout to prevent overlapping elements (optional)
    plt.show()

```
